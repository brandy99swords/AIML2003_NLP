```python
import google.generativeai as genai
from google.colab import userdata
import time

# 1. Setup
api_key = userdata.get('GEMINI_API_KEY')
genai.configure(api_key=api_key)

# 2. Use the stable 1.5-flash model to bypass the 2.5-flash quota bug
model = genai.GenerativeModel('gemini-1.5-flash')

for i in range(3):
    try:
        response = model.generate_content("Say hello to AIML 2003!")
        print(response.text)
        break
    except Exception as e:
        print(f"Attempt {i+1} failed: {e}")
        time.sleep(2) # Brief pause to reset the rate limit
```
